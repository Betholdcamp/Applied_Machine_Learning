```python
#PROBLEM 6
# A. with Sklearn

from sklearn.datasets import make_blobs
f_X_a, f_y_a = make_blobs(n_samples=250, centers=[-1,-1],  n_features=2, cluster_std=1/8, random_state=42) 
f_X_b, f_y_b = make_blobs(n_samples=250, centers=[1,1], n_features=2, cluster_std=1/8, random_state=42)
plt.scatter(f_X_a[:,0], f_X_a[:,1], c='b', label='Class A', alpha=0.2)
plt.scatter(f_X_b[:,0], f_X_b[:,1], c='r', label='Class B', alpha=0.2)
plt.legend()
plt.show()
```

```python
# B. With Numpy
np.random.seed(42)
f_num_a = np.random.randn(250, 2)*1/8 + [-1,-1]
f_num_b = np.random.randn(250, 2)*1/8 + [1,1]
plt.scatter(f_num_a[:,0], f_num_a[:,1], c='b', label='Class A', alpha=0.2)
plt.scatter(f_num_b[:,0], f_num_b[:,1], c='r', label='Class B', alpha=0.2)
plt.legend()
plt.title('With Numpy')
plt.show()
```

```python
# C. Standard deviation 4 and 5
from sklearn.datasets import make_blobs
f_X_a, f_y_a = make_blobs(n_samples=250, centers=[-1,-1],  n_features=2, cluster_std=4, random_state=42) 
f_X_b, f_y_b = make_blobs(n_samples=250, centers=[1,1], n_features=2, cluster_std=5, random_state=42)
plt.scatter(f_X_a[:,0], f_X_a[:,1], c='b', label='Class A', alpha=0.2)
plt.scatter(f_X_b[:,0], f_X_b[:,1], c='r', label='Class B', alpha=0.2)
plt.legend()
plt.show()
```

```python
# D. Not linearly separable
from sklearn.datasets import make_circles
f_X_b, f_y_b = make_blobs(n_samples=5000, centers=[0,0], n_features=2, cluster_std=1/8, random_state=42)
f_cir_a, f_cir_y = make_circles(n_samples=500, random_state=42, factor=0.99) 
plt.scatter(f_cir_a[:,0], f_cir_a[:,1], c='r', label='Class B', alpha=0.2)
plt.scatter(f_X_b[:,0], f_X_b[:,1], c='B', label='Class A', alpha=0.2)
plt.legend()
plt.show()
```
